This is a little visualization project that looks at the gender gap in degrees awarded in different majors in the US between 1970 and today.  There is not much to the project beyond the visualization, but there are some interesting (if tentative) conclusions to be drawn, and a number of interesting questions to consider.  Perhaps I will pursue those questions in the future, but different data would be necessary.

This [dataset](http://www.randalolson.com/2014/06/14/percentage-of-bachelors-degrees-conferred-to-women-by-major-1970-2012/) is derived from annual results released by The Department of Education statistics, which was cleaned and prepared by Randal Olsen from the University of Pennsylvania.

And here is a plot of all 17 of the major categories in a grid, allowing comparison of percentage of men and women in the various fields.  Categories have been prearranged according to percent of women graduating in that field and are plotted vertically along the grid.  

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_grey = (171/255, 171/255, 171/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

degree_cats = [stem_cats, lib_arts_cats, other_cats]
category = -1

fig = plt.figure(figsize=(16, 16))

for sp in range(0,18):
    ax = fig.add_subplot(6, 3, sp+1)
    
    # cycle through the category categories
    category = (category + 1) % 3
    degree = degree_cats[category]
    try:
        ax.plot(women_degrees['Year'], women_degrees[degree[int(sp/3)]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[degree[int(sp/3)]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(degree[int(sp/3)])
    except IndexError:
        # catching empty plot #17
        ax.set_yticklabels([])
    
    # do for all subplots
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.set_yticks([0, 100])
    if sp != 16:
        ax.axhline(50, c=cb_grey, alpha=0.3)
    
    # putting x-axis labels back on the bottom plots
    if sp in [13, 15, 17]:
        ax.tick_params(labelbottom='on')
    
    
    # text annotations
    if sp == 0:
        ax.text(2002, 87, 'Women')
        ax.text(2005, 8, 'Men')
    if sp == 1:
        ax.text(2002, 80, 'Women')
        ax.text(2005, 9, 'Men')
    if sp == 2:
        ax.text(2002, 92, 'Women')
        ax.text(2005, 0, 'Men')
    if sp == 15:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 4, 'Women')
    if sp == 17:
        ax.text(2005, 65, 'Men')
        ax.text(2002, 29, 'Women')
        
# for some reason fig.savefig exports a blank image
plt.savefig('gender_degrees.png')    
plt.show()
```

Some points of note and questions for future projects.  Note that I am writing as if the data is more current than it actually is.  A proper study would need to bridge the year gap between 2011 and today.

## Gender Gap in STEM

The gender gap in the more math and engineering focused STEM fields  (so, everything in the first column save Psychology and Biology) is still significant, and actually has grown wider in Mathematics, Physics and Computer Science.  Both of these areas experienced a peak in terms of gender parity in the 80's, but have regressed somewhat.  In the case of Computer Science this regression is quite pronounced.  However, Psychology and Biology are actually being awarded to many more women than men.  This, combined with the expansion of computation into diverse fields, including Biology, Pychology, the Social Sciences and the Health industry, may account for much of that loss in pure Computer Science degrees.   

This could form an interesting kernel for further research:  although a smaller portion of graduates in Computer Science are women now than at any time since the early 70's, what portion of graduates in computationally heavy fields are women?  We would need more a fine grained parsing of majors, so that we could count graduates in fields like Computational Biology, Genetics and Particle Physics separately from, say, Zoology.   

One small surprise to me us that the gap in Math and Statistics is so small, given the general narrative (which I don't dispute) of women not being involved enough in STEM.  This bears looking into further as well, and hints again at the benefits that a more fine-grained categorization of major could yield.


## The Gapless Fields

The Social Sciences, Business and Agriculture are currently the fields with the _least_ gender gap, according to our visualization.  The split is very near 50-50 in each one.  Is there something special about these fields?  Social Sciences has been one of the more egalitarian (with respect to gender) majors since 1970, but Business and Agriculture went from being boy's clubs to being, well, boy's and girl's clubs.  I wonder about degrees in [Turf Management](https://www.worldcampus.psu.edu/degrees-and-certificates/turfgrass-management-certificate/overview), though...



## The Gaps We Don't Mind So Much


And then there are the other gaps:  those in majors that have more _women_ graduates than men.  Probably on the whole these fields pay a bit less, if only because of the gender-parity issues with regard to pay that we still experience today.  But it does seem that when talking about these gender-gap issues, we emphasize the fields where women are underrepresented, deemphasizing those where they form a majority of graduates.  And while the value inherent in Public Administration, Education and the Health professions is perhaps harder to gage by strictly economic metrics than the value STEM fields, I for one would not be willing to discount them as heavily as a purely economic analysis might do.  Basically, I am saying that maybe things aren't looking so bad for gender-parity, and maybe a 50-50 split straight down the line for all majors is not feasible, nor even desireable.  Of course, there is still much area for concern, and this data doesn't even consider the gender gap in many graduate programs, which I have heard is even greater than the gap in undergraduate degrees in many instances.
